```python
import cv2
from google.colab.patches import cv2_imshow

# Input and output video file paths
input_path = '/content/test.mp4'   # Upload your video to Colab
output_path = 'output_greyscale.avi'

# Open the video file
cap = cv2.VideoCapture(input_path)

# Get video properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# Define codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height), isColor=False)

frame_count = 0

# Process frames
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Convert frame to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Write frame to output
    out.write(gray_frame)

    # Show only first few frames as preview (optional)
    if frame_count < 5:
        print(f"Frame {frame_count + 1}")
        cv2_imshow(gray_frame)
    frame_count += 1

# Release resources
cap.release()
out.release()
print("Grayscale video saved as:", output_path)
```
