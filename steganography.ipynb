```python
import random
import string

def random_string(length=10):
    letters = string.ascii_letters + string.digits
    return ''.join(random.choice(letters) for _ in range(length))

filename = 'random_strings.txt'
with open(filename, 'w') as f:
    for _ in range(1000):
        f.write(random_string() + '\n')
```
